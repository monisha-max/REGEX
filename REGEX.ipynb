```python
import random
import string
import csv

first_names=["Monisha","Mahathi","Sahasra","isha","bhuvana","harshitha","niharika","aara","aadya","aura","john","noah","ava"]
middle_initials=list(string.ascii_uppercase)
family_names=["kollipara","mereddy","marepally","kuruva","miller","reddy","chowdhary","dhary","ashotosh","ilson","erson","aira"]

courses_with_credits={
    "CS3126": 3,
    "CS3202": 4,
    "CS3310": 3,
    "CS3421": 2,
    "CS3515": 3,
    "CS3620": 4
}

sections = {
    "AI1": "@ai1.mahindrauniversity.edu.in",
    "AI2": "@ai2.mahindrauniversity.edu.in",
    "AI3": "@ai3.mahindrauniversity.edu.in",
    "AI4": "@ai4.mahindrauniversity.edu.in",
    "AI5": "@ai5.mahindrauniversity.edu.in"
}

def generating_studentname():
    first_name = random.choice(first_names)
    middle_initial = random.choice(middle_initials)
    family_name = random.choice(family_names)
    return f"{first_name} {middle_initial}. {family_name}"

def generating_rollnumber(batch_code, sequence):
    return f"{batch_code}{str(sequence).zfill(3)}"

def generating_courses():
    course_list = list(courses_with_credits.keys())
    selected_courses = random.sample(course_list, random.randint(3, 5))
    total_credits = sum(courses_with_credits[course] for course in selected_courses)
    return selected_courses, total_credits

def generating_emails(name, section):
    email_username = "".join(name.lower().split())
    email_domain = sections[section]
    return f"{email_username}{email_domain}"

def assigning_sections_to_students(roll_number):
    last_digit = int(roll_number[-1])
    if last_digit < 2:
        return "AI1"
    elif last_digit < 4:
        return "AI2"
    elif last_digit < 6:
        return "AI3"
    elif last_digit < 8:
        return "AI4"
    else:
        return "AI5"

def generating_student_records(num_records):
    records = []
    batch_code = "SE22UARI"
    for i in range(1, num_records + 1):
        name = generating_studentname()
        roll_number = generating_rollnumber(batch_code, i)
        section = assigning_sections_to_students(roll_number)
        courses, total_credits = generating_courses()
        email = generating_emails(name, section)
        records.append([name, roll_number, ",".join(courses),total_credits,email,section])
    return records

student_records = generating_student_records(100000)

with open("student_records.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Student Name", "Roll Number", "Courses Taken", "Total Credits", "Email", "Section"])
    writer.writerows(student_records)

print("Dataset generation complete")
```

```python
import re
import csv
import pandas as pd
from IPython.display import display

def match_roll_number(roll_number):
    pattern = r"^SE22UARI\d{3}$"
    return re.match(pattern, roll_number)

def match_email(email):
    pattern = r"^[a-z0-9]+@(ai1|ai2|ai3|ai4|ai5)\.mahindrauniversity\.edu\.in$"
    return re.match(pattern, email)

def match_course(courses, course_code):
    pattern = re.compile(fr"\b{course_code}\b")
    return pattern.search(courses)

def predefined_search():
    print("Predefined Search Options:")
    print("1)Find all students with Roll Numbers")
    print("2)Find all students with university emails")
    print("3)Find all students enrolled in a specific course")

    choice = input("Enter choice(1/2/3):")
    data = []

    if choice == "1":
        with open("student_records.csv","r") as file:
            reader=csv.reader(file)
            next(reader)
            for row in reader:
                if match_roll_number(row[1]):
                    data.append([row[0],row[1]])
        df=pd.DataFrame(data,columns=["Student Name","Roll Number"])
        display(df)

    elif choice == "2":
        with open("student_records.csv","r") as file:
            reader =csv.reader(file)
            next(reader)
            for row in reader:
                if match_email(row[4]):
                    data.append([row[0], row[4]])
        df = pd.DataFrame(data, columns=["Student Name","University Email"])
        display(df)

    elif choice == "3":
        course_code = input("Enter the course code you'd like to search for (e.g,CS3202): ")
        with open("student_records.csv","r") as file:
            reader = csv.reader(file)
            next(reader)
            for row in reader:
                if match_course(row[2],course_code):
                    data.append([row[0],row[2],row[3]])
        df =pd.DataFrame(data, columns=["Student Name", "Courses","Total Credits"])
        display(df)

    else:
        print("invalid choice")

predefined_search()
```
